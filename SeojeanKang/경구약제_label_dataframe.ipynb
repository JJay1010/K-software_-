```python
import os

# 레이블 json 파일 경로 리스트로 만들기

filepath = 'druglabel/'

labeldir1 = [os.path.join(filepath, i) for i in os.listdir(filepath) if os.path.isdir(os.path.join(filepath, i))]

labeldir2=[]
for i in labeldir1:
  fdir = os.listdir(i)
  for j in fdir:
    if os.path.isdir(os.path.join(i, j)):
      labeldir2.append(i + '/' + j)

labeldir3 = []
for i in labeldir2:
  fdir = os.listdir(i)
  for j in fdir:
    labeldir3.append(i + '/' + j)

len(labeldir3)
```

```python
# json 파일을 DataFrame으로 변환

import json
import pandas as pd
pd.set_option('display.max_columns', None)

# label1 = pd.read_json(first_100[0])
# label1

# pd.read_json / pd.json_normalize / 
# for문으로 json 열고, append나 concat으로 데이터프레임

# 첫번째 json 데이터프레임으로 만들기
with open(labeldir3[0]) as f:
    js = json.loads(f.read())

jsonlabel = pd.json_normalize(js['images'])
jsonlabel
```

```python
a = len(labeldir3) // 100

for j in range(a):
  for i in labeldir3[j * 100 +1:(j+1)*100 + 1]:
    with open(i, encoding='utf-8') as f:
      try:
        js2 = json.loads(f.read()) # dict
        js2 = pd.json_normalize(js2['images'])
        jsonlabel = jsonlabel.append(js2, ignore_index=True)
      except UnicodeDecodeError:
        continue
```
